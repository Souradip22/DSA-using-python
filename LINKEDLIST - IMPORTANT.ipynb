```python
class Node:
    

    def __init__(self, data):
        self.data = data
        self.next = None

        
class LL:
    def __init__(self):
        self.head = None
        self.size = 0
     
    # insert element at first
    def insertFirst(self, val):
        newNode = Node(val)
        self.size += 1
        if self.head == None:        
            self.head = newNode
            return self.head

        newNode.next = self.head
        self.head = newNode
        
        return self.head
    
    def insertLast(self, val):
        newNode = Node(val)
        self.size += 1
        temp = self.head
        if temp == None:
            self.head = newNode
            return self.head
            
        while temp.next != None:
            temp = temp.next
        
        temp.next = newNode
        temp = newNode
        
        return self.head
    
    def insertAtI(self, index, val):
        if index < 0 or index > self.size:
            print('Index out of range : ', index)
            return self.head
        
        
        if index == 0:
            return self.insertFirst(val)
        
        if index == (self.size ):
            return self.insertLast(val)
            
        prev = self.head
        j = 0
        while j < (index - 1):
            prev = prev.next
            j += 1
            
        newNode = Node(val)
        newNode.next = prev.next
        prev.next = newNode
        
        return self.head
    
    
    def deleteFirst(self):
        if self.head == None:
            return None
        
        self.head = self.head.next
        self.size -= 1
        
        return self.head
    
    def deleteLast(self):
        
        if self.head == None:
            return None
        
        temp = self.head
        prev = None
        while temp.next != None:
            prev = temp
            temp = temp.next

        if prev == None:
            self.head = None
            return self.head
            
        prev.next = None
        self.size -= 1
        
        return self.head
    
    def deleteAtI(self, index):
        if index < 0 or index > self.size:
            print('Index out of range : ', index)
            return self.head
        
        
        if index == 0:
            return self.deleteFirst()
        
        if index == (self.size):
            return self.deleteLast()
        
        temp = self.head
        prev = None
        
        j = 0
        while j < (index ):
            prev = temp
            temp = temp.next
            j += 1

        prev.next = temp.next
        self.size -= 1
        return self.head
    
    def findNode(self, val):
        
        if self.head == None:
            return self.head
        temp = self.head
        while temp != None:
            if temp.data == val:
                return True
            temp = temp.next
        return False
    
    def getNode(self, index):
        
        if self.head == None:
            return self.head
        temp = self.head
        j = 0
        while temp != None:
            if j == index:
                return temp
            temp = temp.next
            j += 1
        return None
    
    
    def insertRecursively(self, val , index):
        self.head = self._helperInsertRecursively(val, index, self.head)
        return self.head
    
    def _helperInsertRecursively(self, val, index, node):
        if index < 0 or index > self.size + 1:
            print('Index out of range : ', index)
            return self.head
        if index == 0:
            newNode = Node(val)
            newNode.next = node
            self.size += 1
            return newNode
        index -= 1
        node.next = self._helperInsertRecursively(val, index,  node.next)
        return node
    
    # Remove duplicates from sorted list
    def removeDup(self):
        temp = self.head
        while temp.next != None:
            
            if temp.data == temp.next.data:
                temp.next = temp.next.next
                self.size -= 1
            else:
                temp = temp.next
        return self.head
           

    def printLL(self):
        temp = self.head
        while temp != None:
            print(str(temp.data) + ' -> ', end = "")
            temp = temp.next
        print('None')
#         print('SIZE : ', self.size)
        return









def printLL(head):
    temp = head
    while temp != None:
        print(str(temp.data) + ' -> ', end = "")
        temp = temp.next
    print('None')
#         print('SIZE : ', self.size)
    return
def mergeTwoList(head1, head2):
    
    res = LL()
    head = res
    
    temp1 = head1
    temp2 = head2
   
    while temp1 != None and temp2 != None:
        
        if temp1.data < temp2.data:
            res.insertLast(temp1.data)
            temp1 = temp1.next
        else:
            res.insertLast(temp2.data)
            temp2 = temp2.next

    while (temp1 != None):
        res.insertLast(temp1.data)
        temp1 = temp1.next
        
    while (temp2 != None):
        res.insertLast(temp2.data)
        temp2 = temp2.next
        
    return head


def mergeTwoList2(head1, head2):
    
    res = None
    head = res
    
    temp1 = head1
    temp2 = head2
   
    while temp1 != None and temp2 != None:
        
        if temp1.data < temp2.data:
            res = append(res, temp1.data)
            temp1 = temp1.next
        else:
            res = append(res, temp2.data)
            temp2 = temp2.next

    while (temp1 != None):
        res = append(temp1.data)
        temp1 = temp1.next
        
    while (temp2 != None):
        res = append(res, temp2.data)
        temp2 = temp2.next
        
    return res

def append(head, data):
    curr = head
    newNode = Node(data)

    if head is None:
        head = newNode
        return head

    while curr.next != None:
        curr = curr.next

    curr.next = newNode

    return head


def hasCycle(self, head):

    if head == None:
        return head

    slow = head
    fast = head
    while fast != None and fast.next != None:
        slow = slow.next
        fast = fast.next.next
        #if the Linked List has a cycle slow and fast will meet at some point.
        if slow == fast:
            return True
    return False

def lengthOfCycle(self, head):
    size = 0
    if head == None:
        return head

    slow = head
    fast = head
    while fast != None and fast.next != None:
        slow = slow.next
        fast = fast.next.next
        #if the Linked List has a cycle slow and fast will meet at some point.
        # to calculate the length of cycle, when slow and fast ptr meet
        # loop over the slow pointer only and calulate the size till it matches again with the fast
        if slow == fast:
            while True:  
                slow = slow.next  
                size += 1 
                if slow == fast:  
                    break  
    return size

def detectCycle(self, head):
    if head == None:
        return head

    slow = head
    fast = head
    while fast.next != None and fast.next.next != None:
        slow = slow.next
        fast = fast.next.next
        #if the Linked List has a cycle slow and fast will meet at some point.
        if slow == fast:
            return True
    return False



    


#Take User Input
# ll = takeInput()

newHead = LL()

newHead.insertFirst(10)
newHead.insertFirst(11)
newHead.insertFirst(12)
newHead.insertFirst(13)

newHead.insertLast(21)
newHead.insertLast(22)
newHead.insertLast(23)


newHead.insertAtI(0,31)
newHead.insertAtI(8,32)
newHead.insertAtI(5, 44)


newHead.printLL()
newHead.deleteFirst()
# newHead.deleteFirst()

newHead.deleteLast()
# newHead.deleteLast()

print('Delete At Particular Index')
newHead.deleteAtI(0)
newHead.deleteAtI(3)
newHead.deleteAtI(5)

print('After Delete Operation')
newHead.printLL()
print('Node Present by value : ' ,newHead.findNode(11))

nodeAtI = newHead.getNode(3)
print('Node at i : ', nodeAtI and nodeAtI.data)

newHead.insertRecursively(99,1)



newHead.printLL()
print('-------------------------REMOVE DUPLICATES-----------------------------------')

ll2 = LL()

ll2.insertLast(1)
ll2.insertLast(1)
ll2.insertLast(1)
ll2.insertLast(2)
ll2.insertLast(2)
ll2.insertLast(4)

ll2.printLL()
ll2.removeDup()

ll2.printLL()
print('-------------------------MERGE TWO SORTED LIST-----------------------------------')

ll3 = LL()

ll3.insertLast(1)
ll3.insertLast(1)
ll3.insertLast(3)
ll3.insertLast(5)
ll3.insertLast(7)

ll3.printLL()

ll4 = LL()

ll4.insertLast(2)
ll4.insertLast(4)
ll4.insertLast(6)
ll4.insertLast(8)

ll4.printLL()

res = mergeTwoList(ll3.head, ll4.head)
res.printLL()

res = mergeTwoList2(ll3.head, ll4.head)
printLL(res)


print('-------------------------CYCLE DETECTION-----------------------------------')
ll5 = LL()

ll5.insertLast(1)
ll5.insertLast(1)
ll5.insertLast(3)
ll5.insertLast(5)
ll5.insertLast(7)

# ll5.hasCycle()

```
