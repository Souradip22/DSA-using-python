### Reverse a Linked list

```python
def reverseLinkedList(head):
    if head is None:
        return 
    curr = head
    prev = None
    nextNode = curr.next
    
    while curr != None:
        curr.next = prev
        prev = curr
        curr = nextNode
        if nextNode:
            nextNode = nextNode.next
    head = prev
    return head

arr = [12,3,5,2,1]
L1 = takeInput(arr)

printLL(L1)

L1 = reverseLinkedList(L1)
printLL(L1)
    
        
```

### Detect Loop in Linked List

```python
def hasCycle(head):
    if head is None:
        return False
    slow = head
    fast = head
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False
```

```python
def isHappy(n) -> bool:

    def getSum(n):
        lsum = 0

        while n > 0:
            rem = n % 10
            lsum += rem * rem
            n = n // 10
        return lsum
    
    slow = getSum(n)
    fast = getSum(getSum(n))
    
    while slow != fast and fast != 1:
            slow = getSum(slow)
            fast = getSum(getSum(fast))

    return fast==1

print(isHappy(19))
```

### 83. Remove Duplicates from Sorted List

```python
def deleteDuplicates(head):
    temp = head
    while temp!= None:

        if temp.next and temp.data == temp.next.data:
            temp.next = temp.next.next
        else:
            temp = temp.next
    return head

arr = [1,2,2,4,4,5]
L2 = takeInput(arr)
    
newHead = deleteDuplicates(L2)
printLL(newHead)
```

### Remove Duplicates From an Unsorted Linked List

```python
# APPROACH 1 - O(N2) for each node check all other nodes
# APPROACH 2 - O(NLogN) sort and then apply remove duplicates from a sorted linked list
# APPROACH 3 - O(N) with O(N) extra space HASHMAP

def deleteDuplicates2(head):
    hashSet = set()
    hashSet.add(head.data)
    prev = head
    curr = head.next
    while curr:
        if curr.data in hashSet:
            prev.next = curr.next
        else:
            prev = curr
            hashSet.add(curr.data)
        curr = curr.next
    return head

arr = [1,2,3,5,2,4,4,5]
# arr = [4,2,5,4,2,2]
L2 = takeInput(arr)
    
newHead = deleteDuplicates2(L2)
printLL(newHead)
```

### Remove loop in Linked List

```python
def removeLoop(head):
    
    slow = head
    fast = head

    while slow and fast and fast.next:
        slow = slow.next
        fast = fast.next.next

        if slow == fast:#Loop present
            entry = head
            while slow.next != entry.next:
                slow = slow.next
                entry = entry.next
            slow.next = None
            return head

    return head
```

### 143. Reorder List
    Input: head = [1,2,3,4]
    Output: [1,4,2,3]
    
    L0 → L1 → … → Ln - 1 → Ln ===>>>> L0 → Ln → L1 → Ln - 1 → L2 → Ln - 2 → …


```python
def reorderList(head):
    if head is None:
        return 
    
    slow = head
    fast = head.next
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    # get the middle and reverse from there to end
    secondListStart = slow.next
    secondListEnd = None
    while secondListStart:
        temp = secondListStart.next
        secondListStart.next = secondListEnd
        secondListEnd = secondListStart
        secondListStart = temp

    first = head
    last = secondListEnd
    printLL(first)
    printLL(last)
    #next one by one pick one element from each list
    temp1 = None
    temp2 = None
    
    while first and last:
        if first == last:
            break
        
        temp1 = first.next
        first.next = last
        
        temp2 = last.next
        last.next = temp1
        
        first = temp1
        last = temp2
        
    first.next = None
        
    return head
    
head = [1,2,3,4]
# Output: [1,4,2,3]
LL = takeInput(head)
LL = reorderList(LL)
printLL(LL)
```

### 234. Palindrome Linked List

```python
def isPalindrome(head):
    
    slow = fast = head
    
    #Find the middle Element (will be in slow)
    while fast and fast.next:
        fast = fast.next.next
        slow = slow.next
        
    #reverse the second half also keep a copy to reverse it back
    prev = None
    while slow:
        tmp = slow.next
        slow.next = prev
        prev = slow
        slow = tmp
    copy = prev 
    
    left, right = head, prev
    flag = 0
    while right:
        if right.data != left.data:
            flag = 1
        left = left.next
        right = right.next

    prev = None
    while copy:
        tmp = copy.next
        copy.next = prev
        prev = copy
        copy = tmp
    left = copy

    if flag:
        return False
    return True

arr = [1,2,2,5,1]
ll = takeInput(arr)
printLL(ll)
ans = isPalindrome(ll)
printLL(ll)
print(ans)
```

### 21. Merge Two Sorted Lists

    Input: list1 = [1,2,4], list2 = [1,3,4]
    Output: [1,1,2,3,4,4]

```python
def mergeTwoLists(L1, L2):
    temp = dummyNode = Node(0)
    
    while L1 and L2:
        
        if L1.data < L2.data:
            temp.next = L1
            L1 = L1.next
        else:
            temp.next = L2
            L2 = L2.next
        temp = temp.next
    temp.next = L1 or L2
    return dummyNode.next

list1 = [1,2,3]
list2 = [1,3,4]

L1 = takeInput(list1)
L2 = takeInput(list2)

new = mergeTwoLists(L1, L2)
printLL(new)
```

    It is not possible to reverse a simple singly linked list in less than O(n). A simple singly linked list can only be reversed in O(n) time using recursive and iterative methods. 

    A doubly linked list with head and tail pointers while only requiring swapping the head and tail pointers which require lesser operations than a singly linked list can also not be done in less than O(n) since we need to traverse till the end of the list anyway to find the tail node.

```python
def removeNthFromEnd(head, n):
    if head is None:
        return None
    
    first = dummyNode = Node(0)
    dummyNode.next = head
    second = head
    # take two pointer first and second, first pointer will point to the head and second pointer should point to the n 
    # distance from head
    while n:
        second = second.next
        n -= 1
        
    #now move both the pointer till second reaches to the end as the difference alwasy n then first pointer will 
    # point to the nth element
    while second:
        second = second.next
        first = first.next
        
    if first == dummyNode:
        first.next = first.next.next
        head = first
        head = head.next
    else:
        first.next = first.next.next

    return head
    
# head = [1,2,3,4,5]
# n = 2
head = [1]
n = 1
ll = takeInput(head)
ll = removeNthFromEnd(ll, n)
printLL(ll)

def removeNthFromEnd2(head, n):
    pos = length(head) - n
    print(pos)
    if pos < 0 or pos > length(head) - 1:
        return head

    j = 0
    curr = head
    prev = None
    nextNode = None

    while j < pos:
        prev = curr
        curr = curr.next
        nextNode = curr.next
        j += 1

    if prev is not None:
        prev.next = nextNode
    else:
        head = head.next

    return head

def length(head) :
    count = 0
    curr = head
    while curr is not None:
        count += 1
        curr = curr.next
    return count

head = [1,2,3,4,5]
n = 2
ll = takeInput(head)
ll = removeNthFromEnd2(ll, n)
printLL(ll)
```

### Get Nth From Last Node

```python
def getNthFromLast(head,n):
    #code here
    if head is None:
        return None
    
    first = dummyNode = Node(0)
    dummyNode.next = head
    second = head
    while n and second:
        second = second.next
        n -= 1
    #if n is still present that measn n is greatrer than total number of elements in that case return -1
    if n:
        return -1
    while second:
        second = second.next
        first = first.next

    return first.next.data if first.next else -1
```

### Given a linked list of 0s, 1s and 2s, sort it

```python
def segregate(head):
    zeroHead = zeroTail = oneHead = oneTail = twoHead = twoTail = None
    
    temp = head
    
    while temp:
        
        if temp.data == 0:
            
            if zeroHead is None:
                zeroHead = temp
                zeroTail = temp
            else:
                zeroTail.next = temp
                zeroTail = zeroTail.next
                
        elif temp.data == 1:
            
            if oneHead is None:
                oneHead = temp
                oneTail = temp
            else:
                oneTail.next = temp
                oneTail = oneTail.next
                
        elif temp.data == 2:
            
            if twoHead is None:
                twoHead = temp
                twoTail = temp
            else:
                twoTail.next = temp
                twoTail = twoTail.next
                
        temp = temp.next
    if zeroTail:
        zeroTail.next = oneHead or twoHead
    if oneTail:
        oneTail.next = twoHead
    if twoTail:
        twoTail.next = None
        
    head = zeroHead or oneHead or twoHead
    return head
            
    
head = [1,2,2,1,2,0,2,2]
head = [2, 2, 1, 2, 1]
head = [2, 2, 0, 2, 0, 2, 2, 0, 0, 2, 0]
ll = takeInput(head)
ll = segregate(ll)
printLL(ll)
```

### Intersection of two sorted Linked lists

    Input:
    L1 = 1->2->3->4->6
    L2 = 2->4->6->8
    Output: 2 4 6



```python
def findIntersection(head1,head2):
    
    temp1 = head1
    temp2 = head2
    hashSet = set()
    while temp1:
        hashSet.add(temp1.data)
        temp1 = temp1.next

    dummy = Node(0)
    temp3 = dummy
    while temp2:
        
        if temp2.data in hashSet:
            temp3.next = Node(temp2.data)
            temp3 = temp3.next
            hashSet.remove(temp2.data)
            
        temp2 = temp2.next
        
    return dummy.next
   

head1 = [1,2,3,4,6]
head2 = [2,4,6]

head1 = [1, 3, 7, 8, 9]
head2 = [3, 5,9, 9, 13, 14, 16]
ll1 = takeInput(head1)
ll2 = takeInput(head2)

ll = findIntersection(ll1, ll2)
printLL(ll)
    
```

### 60. Intersection of Two Linked Lists
    A+B AND B+A
    The traversal is contiguous and interchanged, it gives illusion of concatenation, all it does is making sure most pointers travel equal distance before hitting the node of intersection

```python
def getIntersectionNode(headA, headB):
    if headA is None or headB is None:
        return None
    
    a = headA
    b = headB
    
    while a != b:
        
        a = headB if not a else a.next
        b = headA if not b else b.next
    return a.data if a else -1

A = [4,1]
B = [5,6,1]
C = [8,4,5]

ll1 = takeInput(A)
ll2 = takeInput(B)
ll3 = takeInput(C)

t1 = ll1
while t1.next:
    t1 = t1.next
    
t1.next = ll3

t2 = ll2
while t2.next:
    t2 = t2.next

t2.next = ll3

printLL(ll1)
printLL(ll2)

node = getIntersectionNode(ll1, ll2)
print(node.data)
```

### 24. Swap Nodes in Pairs
    Input: head = [1,2,3,4]
    Output: [2,1,4,3]
    
Easy way to solve this is to draw each step

```python
def swapPairs(head):
    prev = dummy = Node(0)
    dummy.next = head

    curr = head

    while curr:
        #store the current next
        adv = curr.next
        nextPtr = adv and adv.next
        
        #Reverse the linkanges
        if adv:
            adv.next = curr
            curr.next = nextPtr
            prev.next = adv
            
        # update the links
        prev = curr
        curr = nextPtr
    return dummy.next

arr = [1,2,3,4]
l1 = takeInput(arr)

head = swapPairs(l1)
printLL(head)
```

### 445. Add Two Numbers II
    Input: l1 = [2,4,3], l2 = [5,6,4]
    Output: [7,0,8]

```python
def getNum(head):
    lsum = 0
    curr = head
    while curr:
        lsum *= 10
        lsum += curr.data 
        curr = curr.next
    return lsum

def addTwoNums(head1, head2):
    n1 = getNum(head1)
    n2 = getNum(head2)
    
    x = [int(a) for a in str(n1 + n2)]
    
    newHead = None
    for el in range(len(x)-1, -1, -1):
        newNode = Node(x[el])
        if newHead is None:
            newHead = newNode
            temp = newHead
        else:
            temp.next = newNode
            temp = temp.next
            
    return newHead

l1 = [2,4,3]
l2 = [5,6,4]


ll1 = takeInput(l1)
ll2 = takeInput(l2)

head = addTwoNums(ll1, ll2)
printLL(head)
```

### 23. Merge k Sorted Lists

    Input: lists = [[1,4,5],[1,3,4],[2,6]]
    Output: [1,1,2,3,4,4,5,6]
    Explanation: The linked-lists are:
    [
      1->4->5,
      1->3->4,
      2->6
    ]
    merging them into one sorted list:
    1->1->2->3->4->4->5->6

```python
def mergeKLists(lists):

    def mergeList(L1, L2):

        temp = dummyNode = Node(0)

        while L1 and L2:
            if L1.data < L2.data:
                temp.next = L1
                L1 = L1.next
            else:
                temp.next = L2
                L2 = L2.next
            temp = temp.next

        temp.next = L1 or L2
        return dummyNode.next


    if lists is None or len(lists) == 0:
        return None

    while len(lists) > 1:
        mergedList = []
        for i in range(0, len(lists), 2):
            list1 = lists[i]
            list2 = lists[i + 1] if (i + 1) < len(lists) else None
            mergedList.append(mergeList(list1, list2))
        lists = mergedList
    return lists[0]


# lists = [[1,4,5],[1,3,4],[2,6]]
ll1 = takeInput([1,4,5])
ll2 = takeInput([1,3,4])
ll3 = takeInput([2,6])

lists = [ll1, ll2, ll3]
ll = mergeKLists(lists)
printLL(ll)
```

### Given a sorted doubly linked list of positive distinct elements, the task is to find pairs in a doubly-linked list whose sum is equal to given value x, without using any extra space? 

    Example:  

    Input : head : 1 <-> 2 <-> 4 <-> 5 <-> 6 <-> 8 <-> 9
            x = 7
    Output: (6, 1), (5,2)

```python
def findPair(head, k):
    
    # Write your code here.
    # Return boolean true or false.
    first = head
    second = head
    flag = 0
    
    while second.next != None:
        second = second.next
        
    while first != second and second.next != first:
        
        if (first.data + second.data) == k:
            flag = 1
            first = first.next
            second = second.prev
        else:
            if (first.data + second.data) < k:
                first = first.next
            else:
                second = second.prev
    if flag == 0:
        return False
    return True
```

### 61. Rotate List

    Input: head = [1,2,3,4,5], k = 2
    Output: [4,5,1,2,3]

```python
def rotateRight(head, k):
    if not head:
        return head    
    length, tail = 1, head
    
    while tail.next:
        tail = tail.next
        length += 1
        
    k = k % length
    if k == 0:
        return head
    
    curr = head
    for i in range(length - k - 1):
        curr = curr.next
    
    newHead = curr.next
    curr.next = None
    tail.next = head
    return newHead
        

head = [1,2,3,4,5]
k = 2
ll1 = takeInput(head)
ll1 = rotateRight(ll1, k)
printLL(ll1)
```


### 138. Copy List with Random Pointer
Input: head = [[7,null],[13,0],[11,4],[10,2],[1,0]]
Output: [[7,null],[13,0],[11,4],[10,2],[1,0]]

```python
def copyRandomList(head):

    hashMap = {None:None}

    curr = head
    while curr:
        newNode = Node(curr.val)
        hashMap[curr] = newNode
        curr = curr.next

    curr = head
    while curr:
        copied = hashMap[curr]
        copied.next = hashMap[curr.next]
        copied.random = hashMap[curr.random]

        curr = curr.next

    return hashMap[head]
```

### 86. Partition List
    Given the head of a linked list and a value x, partition it such that all nodes less than x come before nodes greater than or equal to x.

    You should preserve the original relative order of the nodes in each of the two partitions.

    Input: head = [1,4,3,2,5,2], x = 3
    Output: [1,2,2,4,3,5]

```python
def partition(head, x):
        
    ltXHead = ltXTail = gtXHead = gtXTail = None

    curr = head

    while curr:

        if curr.data < x:

            if ltXHead is None:
                ltXHead = curr
                ltXTail = curr
            else:
                ltXTail.next = curr
                ltXTail = ltXTail.next

        else:

            if gtXHead is None:
                gtXHead = curr
                gtXTail = curr
            else:
                gtXTail.next = curr
                gtXTail = gtXTail.next

        curr = curr.next

    if ltXTail:
        ltXTail.next = gtXHead
    if gtXTail:
        gtXTail.next = None

    return ltXHead or gtXHead

head = [1,4,3,2,5,2]
x = 3
ll1 = takeInput(head)
ll1 = partition(ll1, x)
printLL(ll1)
```
