*** If row or column becomes 1, that means in this case we got a potential path, becasue if any one of row or column is 1 then there is always way to go to the destination [Provided allowed to go only R and D] direction. ***

```python
def mazeCountPath(r,c):
    if r ==1 or c == 1:
        return 1
    
    left = mazeCountPath(r - 1 , c)
    right = mazeCountPath(r, c - 1)
    
    return left + right

mazeCountPath(3, 3)
#Assuming start (3, 3)  and need to go to (1, 1)
```

```python
def mazePrintAllPath(p, maze, r, c):
    ans = []
    if r == len(maze) - 1 and  c == len(maze[0]) - 1:
        ans = []
        ans.append(p)
        return ans
    
    if not maze[r][c]:
        return []
    
    maze[r][c] = False
    
    list1 = []

    if r < len(maze) - 1 :
        list1.extend(mazePrintAllPath(p + 'D', maze, r + 1 , c))

    if c < len(maze[0]) - 1:
        list1.extend(mazePrintAllPath(p + 'R' , maze, r, c + 1))
        
    if r > 0:
        list1.extend(mazePrintAllPath(p + 'U' , maze, r - 1, c ))
        
    if c > 0:
        list1.extend(mazePrintAllPath(p + 'L' , maze, r, c - 1))
        
    maze[r][c] = True
    
    return list1
    
    
maze = [[1, 1, 1], [1, 1, 1], [1, 1, 1]]
print(mazePrintAllPath('', maze, 0, 0 ))
```

```python
count = 0
def mazePrintAllPath2(p, maze, r, c, path, step):
    global count 
    

    if r == len(maze) - 1 and  c == len(maze[0]) - 1:
        
        path[r][c] = step
        ans = []
        ans.append(p)
        for arr in path:
            print(arr)
        print(p)
        print()
        count += 1
        return ans
    
    if not maze[r][c]:
        return []
    
    #needed for backtracking
    maze[r][c] = False
    path[r][c] = step
    
    list1 = []
    if r < len(maze) - 1 :
        
        list1.extend(mazePrintAllPath2(p + 'D', maze, r + 1 , c, path, step  + 1))

    if c < len(maze[0]) - 1:
        list1.extend(mazePrintAllPath2(p + 'R' , maze, r, c + 1, path, step + 1))
        
    if r > 0:
        list1.extend(mazePrintAllPath2(p + 'U' , maze, r - 1, c, path, step + 1))
        
    if c > 0:
        list1.extend(mazePrintAllPath2(p + 'L' , maze, r, c - 1, path, step + 1))
    
    #needed for backtracking
    maze[r][c] = True
    path[r][c] = 0

    
    return list1

r, c = (5,5)
maze = [[ True for i in range(c) ] for j in range(r)]
path = [[ 0 for i in range(len(maze[0])) ] for j in range(len(maze))]


print(mazePrintAllPath2('', maze, 0, 0, path, 1 ))
print(count)
```
