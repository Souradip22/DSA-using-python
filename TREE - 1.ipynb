```python
class BinaryTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def takeInput():
    rootData = int(input())
    if rootData == -1:
        return None
    rootNode = BinaryTreeNode(rootData)
    leftNode = takeInput()
    rightNode = takeInput()
    
    rootNode.left = leftNode
    rootNode.right = rightNode
    return rootNode

def printBinaryTree(root):
    
    if root == None:
        return
    print(root.data, end = ' ')
    if root.left:
        print("L", root.left.data, end = ':')
    if root.right:
        print("R", root.right.data, end = ' ')
    print()
    printBinaryTree(root.left)
    printBinaryTree(root.right)
    
# number of nodes in BT

def numberOfNodes(root):
    if root == None:
        return 0
    leftNodeNum = numberOfNodes(root.left)
    rightNodeNum = numberOfNodes(root.right)
    
    return 1 + leftNodeNum + rightNodeNum

# Node with largest data
def largestNode(root):
    if root is None:
        return -1 # ideally it should be - inf
    leftNode = largestNode(root.left)
    rightNode = largestNode(root.right)
    
    return max(root.data, leftNode, rightNode)

def printNodeDepth(root, k, d = 0):
    
    if root == None:
        return
    if d == k:
        print(root.data)
    printNodeDepth(root.left, k, d + 1)
    printNodeDepth(root.right, k, d + 1)
    
    
    
def treeHeight(root):
    if root == None:
        return -1
    lDepth = treeHeight(root.left)
    rDepth = treeHeight(root.right)
    if (lDepth > rDepth):
        return lDepth+1
    else:
        return rDepth+1
    
    
def numberOfLeafNode(root):
    if root == None:
        return 0
    if root.left == None and root.right == None:
        return 1
    numLeafLeft = numberOfLeafNode(root.left)
    numLeafRight = numberOfLeafNode(root.right)
    
    return numLeafLeft + numLeafRight

def changeToDepthTree(root) :

    return _helper(root, d = 0)

def _helper(root, d):
    
    if root == None:
        return
    
    root.data = d
    
    _helper(root.left, d+1)
    _helper(root.right, d+ 1)
    
    
# Replace Node With Depth
# Is Node Present?
# Nodes without sibling
# Nodes Greater Than X
```
