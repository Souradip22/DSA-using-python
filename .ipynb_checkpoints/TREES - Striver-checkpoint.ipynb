```python
#Maximum Depth Of A Binary Tree
def findMaxDepth(root):
    # Write your Code here.
    if root is None:
        return 0

    left = findMaxDepth(root.left)
    right = findMaxDepth(root.right)

    return 1 + max(left, right)
```

```python
# Checked BT is balanced or not while calculating the hight of BT
def isBalanced(root):

    ans = self._helperFun(root)

    if ans == -1:
        return False
    return True

def _helperFun(self, root):

    if root is None:
        return 0

    leftH = self._helperFun(root.left)
    rightH = self._helperFun(root.right)

    if leftH == -1 or rightH == -1:
        return -1

    if abs(leftH - rightH) > 1:
        return -1

    return 1 + max(leftH, rightH)
```

### 124. Binary Tree Maximum Path Sum(HARD)

```python
class Solution:
    def maxPathSum(self, root) -> int:
        if root is None:
            return 0
        ans = [root.val]
        self._helperMaxSum(root, ans)
        return ans[0]

    def _helperMaxSum(self, root, ans):

        if root is None:
            return 0

        left = self._helperMaxSum(root.left, ans)
        right = self._helperMaxSum(root.right, ans)
        if left < 0:
            left =  0
        if right < 0:
            right = 0

        ans[0] = max(ans[0], root.val + left + right)

        return root.val + max(left, right)
```

## Bottom View

```python
def bottomView(self, root):
    if root is None:
        return []

    q = [(root, 0)]
    node_dict = {}

    while q:
        node, hd = q.pop(0)
        node_dict[hd] = node.data

        if node.left:
            q.append((node.left, hd - 1))

        if node.right:
            q.append((node.right, hd + 1))

    bottom_view = [node_dict[hd] for hd in sorted(node_dict)]
    return bottom_view
```

### 101. Symmetric Tree


```python
def isSymmetric(self, root) -> bool:
    if root is None:
        return

    return self._helperSymm(root.left, root.right)

def _helperSymm(self, root1, root2):

    if root1 is None or root2 is None:
        return root1 == root2
    if root1 is None and root2 is None:
        return True

    if root1.val != root2.val:
        return False

    return self._helperSymm(root1.left, root2.right) and self._helperSymm(root1.right, root2.left)
```

### 662. Maximum Width of Binary Tree


```python
 # Calculate width considering the null 

def widthOfBinaryTree(self, root) -> int: 

        maxWidth = 0
        if root is None:
            return 0

        q = queue.Queue()

        q.put((root, 0))

        while not q.empty():

            lmin = float('inf')
            lmax = float('-inf')

            for _ in range(q.qsize()):

                tupl = q.get()

                node = tupl[0]
                ind = tupl[1]

                lmin = min(lmin, ind)
                lmax = max(lmax, ind)

                if node.left:
                    ind1 = 2 * (ind - lmin) + 1
                    q.put((node.left, ind1))

                if node.right:
                    ind2 = 2 * (ind - lmin) + 2
                    q.put((node.right, ind2))
                
            maxWidth = max(maxWidth, lmax - lmin + 1)

        return maxWidth
```

```python
# Null nodes are excluded from the calculation (codeStudio)
import queue
def getMaxWidth(root):
    maxWidth = 0
    if root is None:
        return 0

    q = queue.Queue()
    q.put((root, 0))

    while not q.empty():
      level_size = q.qsize()
      lmin = float('inf')

      for i in range(level_size):
          tupl = q.get()
          node = tupl[0]
          ind = tupl[1]

          lmin = min(lmin, ind)

          if node.left:
              ind1 = 2 * (ind - lmin)
              q.put((node.left, ind1))

          if node.right:
              ind2 = 2 * (ind - lmin) + 1
              q.put((node.right, ind2))

      lmax = lmin + level_size - 1
      maxWidth = max(maxWidth, lmax - lmin + 1)

    return maxWidth
```

### Children Sum Property

```python
def changeTree(root):
    
    if root is None:
        return None
    
    childSum = 0
    if root.left:
        childSum += root.left.data
    if root.right:
        childSum += root.right.data
        
    if childSum > root.data:
        root.data = childSum
        
    else:
        if root.left:
            root.left.data = root.data
        if root.right:
            root.right.data = root.data
    
    changeTree(root.left)
    changeTree(root.right)
    
    tmp = 0
    if root.left:
        tmp += root.left.data
    if root.right:
        tmp += root.right.data
        
    if tmp > root.data:
        root.data = tmp
```

### 863. All Nodes Distance K in Binary Tree (HARD)


```python
from  queue import Queue
class Solution:
    def distanceK( root: TreeNode, target: TreeNode, k: int) -> List[int]:
        if root is None:
            return None

        q = Queue()
        hashMap = {}
        q.put(root)
        
        while not q.empty():
            
            node = q.get()
            
            if node.left:
                hashMap[node.left.val] = node
                q.put(node.left)

            if node.right:
                hashMap[node.right.val] = node
                q.put(node.right)

        q2 = Queue()
        visited = set()
        dist = 0
        
        q2.put(target)
        visited.add(target)

        while not q2.empty():
            
            if dist == k:
                break

            for _ in range(q2.qsize()):
                
                node = q2.get()

                up = hashMap.get(node.val)
                if up and up not in visited:
                    visited.add(up)
                    q2.put(up)

                if node.left and node.left not in visited:
                    visited.add(node.left)
                    q2.put(node.left)
                
                if node.right and node.right not in visited:
                    visited.add(node.right)
                    q2.put(node.right)
            
            dist += 1
        
        ans = []
        while not q2.empty():
            tmp = q2.get()
            ans.append(tmp.val)

        return ans
```

```python
from sys import stdin,setrecursionlimit
from queue import Queue

setrecursionlimit(10**7)

# Binary tree node class for reference.
class BinaryTreeNode :
	def __init__(self, data) :
		self.data = data
		self.left = None
		self.right = None



def timeToBurnTree(root, start):
    if root is None:
        return None
    # Write your code here.
    q1 = Queue()
    hashMap = {}

    q1.put(root)
    startNode = root

    while not q1.empty():

        node = q1.get()
        if node.data == start:
            startNode = node
        if node.left:
            hashMap[node.left.data] = node
            q1.put(node.left)

        if node.right:
            hashMap[node.right.data] = node
            q1.put(node.right)
    
    q2 = Queue()
    q2.put(startNode)
    visited = set()
    visited.add(startNode)
    ans = 0
    while not q2.empty():

        for _ in range(q2.qsize()):

            node = q2.get()

            up = hashMap.get(node.data, None)

            if up and up not in visited:
                q2.put(up)
                visited.add(up)
            
            if node.left and node.left not in visited:
                q2.put(node.left)
                visited.add(node.left)
            
            if node.right and node.right not in visited:
                q2.put(node.right)
                visited.add(node.right)
        ans += 1
    
    return ans - 1

```

### 222. Count Complete Tree Nodes


```python
def countNodes(root):
    # write your code here

    if root is None:
        return 0

    lh = lhs(root)
    rh = rhs(root)

    if lh == rh:
        return int(pow(2, lh) - 1)
    return 1 + countNodes(root.left) + countNodes(root.right)


def lhs(root):
    ans = 0
    while root:
        root = root.left
        ans += 1
    return ans

def rhs(root):
    ans = 0
    while root:
        root = root.right
        ans += 1
    return ans
```

### 106. Construct Binary Tree from Inorder and Postorder Traversal

```python
class Solution:
    def buildTree(self, inorder: List[int], postorder: List[int]) -> [TreeNode]:

        if len(inorder) == 0:
            return None
        
        rootVal = postorder[len(postorder) - 1]
        rootInd = -1
        i = 0
        while i < len(inorder):
            if inorder[i] == rootVal:
                rootInd = i
                break
            i += 1
        leftSide = self.buildTree(inorder[:rootInd], postorder[:rootInd])
        rightSide = self.buildTree(inorder[rootInd + 1:], postorder[rootInd:len(postorder) - 1])
        rootNode = TreeNode(rootVal)
        rootNode.left = leftSide
        rootNode.right = rightSide
        return rootNode
```

### 105. Construct Binary Tree from Preorder and Inorder Traversal

```python
class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> Optional[TreeNode]:

        if len(preorder) == 0:
            return None

        rootVal = preorder[0]
        rootInd = -1
        i = 0
        while i < len(inorder):
            if inorder[i] == rootVal:
                rootInd = i
                break
            i += 1
        leftSide = self.buildTree(preorder[1:rootInd + 1], inorder[:rootInd])
        rightSide = self.buildTree(preorder[rootInd + 1:], inorder[rootInd + 1: ])
        rootNode = TreeNode(rootVal)
        rootNode.left = leftSide
        rootNode.right = rightSide
        return rootNode
```

### 297. Serialize and Deserialize Binary Tree

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None
import queue
class Codec:

    def serialize(self, root):
        """Encodes a tree to a single string.
        
        :type root: TreeNode
        :rtype: str
        """
        if root is None:
            return ""
        q = queue.Queue()
        q.put(root)
        res = "" + str(root.val) + ','
        while not q.empty():

            for _ in range(q.qsize()):

                node = q.get()

                if node.left:
                    q.put(node.left)
                    res = res + str(node.left.val) + ','
                else:
                    res = res + '#' + ','

                if node.right:
                    q.put(node.right)
                    res = res + str(node.right.val) + ','
                else:
                    res = res + '#' + ','
        return res

        

    def deserialize(self, data):
        """Decodes your encoded data to tree.
        
        :type data: str
        :rtype: TreeNode
        """

        dataList = data.split(',') 
        dataList= dataList[:len(dataList)- 1]

        if len(dataList) == 0:
            return None
        rootNode = TreeNode(int(dataList[0]))
        q2 = queue.Queue()
        q2.put(rootNode)
        i = 1
        while not q2.empty():

            node = q2.get()
            if dataList[i] != '#':
                tmpNode = TreeNode(int(dataList[i]))
                node.left = tmpNode
                q2.put(tmpNode)
            
            i += 1
            if dataList[i] != '#':
                tmpNode = TreeNode(int(dataList[i]))
                node.right = tmpNode
                q2.put(tmpNode)
            
            i += 1
        return rootNode
    



        

# Your Codec object will be instantiated and called as such:
# ser = Codec()
# deser = Codec()
# ans = deser.deserialize(ser.serialize(root))
```

### Flatten Binary Tree to Linked List

```python
def flatten(self, root: Optional[TreeNode]) -> None:
    """
    Do not return anything, modify root in-place instead.
    """
    prev = [None]
    self._helperFunc(root, prev)



def _helperFunc(self, root, prev):
    if root is None:
        return

    self._helperFunc(root.right, prev)
    self._helperFunc(root.left, prev)

    root.right = prev[0]
    root.left = None
    prev[0] = root
```

```python
def findCeil(root, x):
    # Write your code here.
    ans = -1

    while root is not None:

        if root.data >= x:
            ans = root.data
            root = root.left
        else:
            root= root.right
        
    return ans

def floorInBST(root, X):

    ans = -1

    while root is not None:

        if root.data <= X:
            ans = root.data
            root = root.right
        else:
            root= root.left
        
    return ans
```

```python
def validateBST(root):
    
    return _helperBST(root, [float('-inf'), float('inf')])
    
def _helperBST(root, ds):

    if root is None:
        return True

    leftSide = _helperBST(root.left, [ds[0], root.data])
    rightSide = _helperBST(root.right, [root.data, ds[1]])

    if not leftSide or not rightSide:
        return False

    if ds[0] < root.data and root.data < ds[1]:
        return True and (leftSide and rightSide)
    
    return False
```
