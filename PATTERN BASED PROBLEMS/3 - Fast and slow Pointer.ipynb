### 457. Circular Array Loop

```python
def circularArrayLoop(nums):

    for i in range(len(nums)):
        isForward = nums[i] >= 0
        slow, fast = i, i

        # if slow or fast becomes '-1' this means we can't find cycle for this number
        while True:
            # move one step for slow pointer
            slow = findNextIndex(nums, isForward, slow)

            # move one step for fast pointer
            fast = findNextIndex(nums, isForward, fast)
            if fast != -1:
                fast = findNextIndex(nums, isForward, fast)

            if slow == -1 or fast == -1 or slow == fast:
                break

        if slow != -1 and slow == fast:
            return True

    return False


def findNextIndex(nums, isForward, currentIndex):
    direction = nums[currentIndex] >= 0

    if isForward != direction:
        return -1 # changes in direction resturn -1

    nextIndex = (currentIndex + nums[currentIndex]) % len(nums)

    #one element cycle: return -1
    if nextIndex == currentIndex:
        return -1

    return nextIndex


nums = [-2,1,-1,-2,-2]
print(circularArrayLoop(nums))
nums = [2,-1,1,2,2]
print(circularArrayLoop(nums))
nums = [-1,2]
print(circularArrayLoop(nums))
```
